````python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∞–≥—Ä–µ–≥–∞—Ü–∏–∏ —ç–∫—Å–ø–µ—Ä—Ç–Ω—ã—Ö –æ—Ü–µ–Ω–æ–∫
def aggregate_scores(row):
    scores = [row['ex_1'], row['ex_2'], row['ex_3']]

    if len(set(scores)) == 3:
        return None

    modal_score = mode(scores)
    return modal_score[0]
````

<div class="alert alert-warning">
–ö–æ–≥–¥–∞ —è –ø—Ä–æ–≤–µ—Ä—è–ª —Ç–≤–æ–π –ø—Ä–æ–µ–∫—Ç (scipy==1.8.0), —Ç–æ –ø–æ—Å–ª–µ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —ç—Ç–æ–≥–æ –∫–æ–¥–∞ –≤–º–µ—Å—Ç–æ –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ—è–≤–∏–ª–∏—Å—å numpy.array. –ò –¥–∞–ª—å–Ω–µ–π—à–∏–π –∫–æ–¥ –ª–æ–º–∞–ª—Å—è.
    
–í–æ–∑–º–æ–∂–Ω–æ —Å—Ç–æ–∏—Ç –≤–æ–∑–≤—Ä–∞—â–∞—Ç—å modal_score[0][0]. –ù–æ –∑–¥–µ—Å—å –≤—Å–µ –º–æ–∂–µ—Ç –∑–∞–≤–∏—Å–µ—Ç—å –æ—Ç –≤–µ—Ä—Å–∏–∏ scipy.
</div>

````python
# –ü—Ä–∏–º–µ–Ω—è–µ–º –∞–≥—Ä–µ–≥–∞—Ü–∏—é
df_expertAnnotations['expert_score'] = df_expertAnnotations.apply(aggregate_scores, axis=1)
# –£–¥–∞–ª—è–µ–º —Å—Ç—Ä–æ–∫–∏, –≥–¥–µ —ç–∫—Å–ø–µ—Ä—Ç—ã —Å–∏–ª—å–Ω–æ —Ä–∞–∑–æ—à–ª–∏—Å—å (–≤–µ—Ä–Ω—É–ª—Å—è None)
df_expertAnnotations = df_expertAnnotations.dropna(subset=['expert_score'])
# –ù–æ—Ä–º–∞–ª–∏–∑—É–µ–º –æ—Ü–µ–Ω–∫—É –≤ –¥–∏–∞–ø–∞–∑–æ–Ω [0, 1] (–¥–µ–ª–µ–Ω–∏–µ–º –Ω–∞ 4)
df_expertAnnotations['expert_score_norm'] = (df_expertAnnotations['expert_score'] - 1) / 3.0
````

````python
# –í—ã—á–∏—Å–ª—è–µ–º –æ–±—â–µ–µ —á–∏—Å–ª–æ –≥–æ–ª–æ—Å–æ–≤
df_crowdAnnotations['total_votes'] = df_crowdAnnotations['n_match'] + df_crowdAnnotations['n_no_match']
# –í—ã—á–∏—Å–ª—è–µ–º "—É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å" –æ—Ü–µ–Ω–∫–∏ (—á–µ–º –±–æ–ª—å—à–µ –≥–æ–ª–æ—Å–æ–≤, —Ç–µ–º —É–≤–µ—Ä–µ–Ω–Ω–µ–µ)
df_crowdAnnotations['confidence'] = 1 - 1/(df_crowdAnnotations['total_votes'] + 1)
# –í–∑–≤–µ—à–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞
df_crowdAnnotations['crowd_score_norm'] = df_crowdAnnotations['desc_perc'] * df_crowdAnnotations['confidence']
````

–û–Ω–∞ –ø—Ä–æ—Å—Ç–æ –∑–∞–Ω–∏–∂–∞–µ—Ç –æ—Ü–µ–Ω–∫—É –∏ –≤—Å—ë.  –í—ã—Å–æ–∫–∏–µ –æ—Ü–µ–Ω–∫–∏ —Å—Ç–∞–Ω—É—Ç –Ω–∏–∂–µ, –∞ –Ω—É–ª–∏ —Ç–∞–∫ –∏ –æ—Å—Ç–∞–Ω—É—Ç—Å—è –Ω—É–ª—è–º–∏.
    
–ö–∞–∫ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤—É –º–æ–∂–Ω–æ —Ä–∞—Å—Å–º–æ—Ç—Ä–µ—Ç—å –∏—Å–∫–ª—é—á–µ–Ω–∏–µ –æ—Ü–µ–Ω–æ–∫, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –º—ã –Ω–µ —É–≤–µ—Ä–µ–Ω—ã –≤ –æ—Ü–µ–Ω–∫–∞—Ö –∫—Ä–∞—É–¥—Å–æ—Ä—Å–µ—Ä–æ–≤.

````python
def merge_annotations(df_expertAnnotations, df_crowdAnnotations):
    merged = pd.merge(
        df_expertAnnotations[['image', 'query_id', 'expert_score_norm']],
        df_crowdAnnotations[['image', 'query_id', 'crowd_score_norm', 'total_votes']],
        on=['image', 'query_id'],
        how='outer'  # –ò—Å–ø–æ–ª—å–∑—É–µ–º outer join –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è –≤—Å–µ—Ö –≤–æ–∑–º–æ–∂–Ω—ã—Ö –ø–∞—Ä
    )

    merged['target'] = np.where(
        merged['expert_score_norm'].notna() & merged['crowd_score_norm'].notna(),
        0.7 * merged['expert_score_norm'] + 0.3 * merged['crowd_score_norm'],
        np.where(
            merged['expert_score_norm'].notna(),
            merged['expert_score_norm'],
            np.where(
                merged['crowd_score_norm'].notna(),
                merged['crowd_score_norm'],
                np.nan
            )
        )
    )

    return merged
````

````python
# –ó–∞–≥—Ä—É–∂–∞–µ–º –º–æ–¥–µ–ª—å –∏ —É–±–∏—Ä–∞–µ–º –ø–æ—Å–ª–µ–¥–Ω–∏–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–æ–Ω–Ω—ã–π —Å–ª–æ–π
resnet = models.resnet18(pretrained=True)
resnet = torch.nn.Sequential(*list(resnet.children())[:-1])  # —É–¥–∞–ª—è–µ–º –ø–æ—Å–ª–µ–¥–Ω–∏–π —Å–ª–æ–π
resnet.eval();  # –ø–µ—Ä–µ–≤–æ–¥–∏–º –≤ —Ä–µ–∂–∏–º –∏–Ω—Ñ–µ—Ä–µ–Ω—Å–∞
````

````python
# –¢—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∞—Ü–∏–∏ —Ç–∞–∫–∏–µ –∂–µ, –∫–∞–∫ –ø—Ä–∏ –æ–±—É—á–µ–Ω–∏–∏ –Ω–∞ ImageNet
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize –¥–æ –Ω—É–∂–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])
````

````python
def get_image_embedding(img_path, model, transform):
    img = Image.open(img_path).convert('RGB')
    img_t = transform(img).unsqueeze(0)  # –¥–æ–±–∞–≤–∏–º —Ä–∞–∑–º–µ—Ä –±–∞—Ç—á–∞
    with torch.no_grad():
        embedding = model(img_t).squeeze().numpy()
    return embedding
````

## 4. –í–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è —Ç–µ–∫—Å—Ç–æ–≤

–°–ª–µ–¥—É—é—â–∏–π —ç—Ç–∞–ø ‚Äî –≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è —Ç–µ–∫—Å—Ç–æ–≤. –í—ã –º–æ–∂–µ—Ç–µ –ø–æ—ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å —Å –Ω–µ—Å–∫–æ–ª—å–∫–∏–º–∏ —Å–ø–æ—Å–æ–±–∞–º–∏ –≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏–∏ —Ç–µ–∫—Å—Ç–æ–≤:

- tf-idf
- word2vec
- \*—Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–µ—Ä—ã (–Ω–∞–ø—Ä–∏–º–µ—Ä Bert)

\* ‚Äî –µ—Å–ª–∏ –≤—ã –∏–∑—É—á–∞–ª–∏ —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–µ—Ä—ã –≤ —Å–ø—Ä–∏–Ω—Ç–µ –ú–∞—à–∏–Ω–Ω–æ–µ –æ–±—É—á–µ–Ω–∏–µ –¥–ª—è —Ç–µ–∫—Å—Ç–æ–≤.


````python
# –ó–∞–≥—Ä—É–∂–∞–µ–º BERT-–º–æ–¥–µ–ª—å –∏ —Ç–æ–∫–µ–Ω–∏–∑–∞—Ç–æ—Ä
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', force_download=True)
bert_model = BertModel.from_pretrained('bert-base-uncased')
bert_model.eval();  # –ø–µ—Ä–µ–≤–æ–¥–∏–º –≤ —Ä–µ–∂–∏–º –∏–Ω—Ñ–µ—Ä–µ–Ω—Å–∞
````

````python
def get_text_embedding(text, tokenizer, model):
    tokens = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=64)
    with torch.no_grad():
        output = model(**tokens)
    # –ü–æ–ª—É—á–∞–µ–º —ç–º–±–µ–¥–¥–∏–Ω–≥–∏ —Ç–æ–∫–µ–Ω–æ–≤ –∏ –¥–µ–ª–∞–µ–º mean pooling
    embeddings = output.last_hidden_state.squeeze(0)
    attention_mask = tokens['attention_mask'].squeeze(0)
    mask_expanded = attention_mask.unsqueeze(-1).expand(embeddings.size()).float()
    mean_pooled = torch.sum(embeddings * mask_expanded, dim=0) / torch.clamp(mask_expanded.sum(dim=0), min=1e-9)
    return mean_pooled.numpy()
````

````python
text_embeddings = {}

for idx, row in tqdm(df_train_filtered.iterrows(), total=len(df_train_filtered)):
    text = row['query_text']
    query_id = row['query_id']

    if query_id not in text_embeddings:
        try:
            emb = get_text_embedding(text, tokenizer, bert_model)
            text_embeddings[query_id] = emb
        except Exception as e:
            print(f'–û—à–∏–±–∫–∞ —Å query_id={query_id}: {e}')
````

````python
def search_image_by_text(
    text_query,
    image_embeddings,
    model,
    tokenizer,
    bert_model,
    top_k: int = 3
) -> dict:
    try:
        text_embedding = get_text_embedding(text_query, tokenizer, bert_model)
        if text_embedding is None:
            return {"error": "–û—à–∏–±–∫–∞ –≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏–∏ —Ç–µ–∫—Å—Ç–∞."}
    except Exception as e:
        return {"error": f"–û—à–∏–±–∫–∞ –æ–±—Ä–∞–±–æ—Ç–∫–∏ —Ç–µ–∫—Å—Ç–∞: {str(e)}"}

    predictions = []
    for img_path, img_emb in image_embeddings.items():
        try:
            combined = np.concatenate([img_emb, text_embedding])
            score = model.predict(combined.reshape(1, -1))[0]
            predictions.append((img_path, float(score)))
        except Exception as e:
            print(f"–û—à–∏–±–∫–∞ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ–º {img_path}: {str(e)}")
            continue

    if not predictions:
        return {"error": "–ù–µ —É–¥–∞–ª–æ—Å—å —Å—Ä–∞–≤–Ω–∏—Ç—å –∑–∞–ø—Ä–æ—Å —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è–º–∏."}

    # 4. –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –∏ –≤—ã–±–æ—Ä –ª—É—á—à–∏—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    predictions.sort(key=lambda x: x[1], reverse=True)

    return {
        "query": text_query,
        "top_matches": predictions[:top_k]
    }
````

````python
def test_search_system(queries, image_embeddings, model, tokenizer, bert_model):
    for query in queries.sample(10)['query_text']:
        result = search_image_by_text(
            text_query=query,
            image_embeddings=image_embeddings,
            model=model,
            tokenizer=tokenizer,
            bert_model=bert_model
        )

        print(f"\n–¢–æ–ø-3 —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –¥–ª—è –∑–∞–ø—Ä–æ—Å–∞: '{query}'")
        for i, (img_path, score) in enumerate(result['top_matches'], 1):
            print(f"{i}. {img_path} (score: {score:.4f})")

            # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ
            try:
                img = Image.open(img_path)
                plt.figure(figsize=(5,5))
                plt.imshow(img)
                plt.title(f"Score: {score:.4f}")
                plt.axis('off')
                plt.show()
            except Exception as e:
                print(f"–ù–µ —É–¥–∞–ª–æ—Å—å –∑–∞–≥—Ä—É–∑–∏—Ç—å {img_path}: {str(e)}")
````
